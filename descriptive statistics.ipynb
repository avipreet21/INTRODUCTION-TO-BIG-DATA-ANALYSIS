```python
df_happiness_2019=df_happiness[df_happiness["Year"]==2019]
```

```python
# drop the two columns Continent and Year

df_happiness_2019=df_happiness_2019.drop(['Continent','Year'], axis=1)
```

```python
# merge the two dataframes based on the common column "Code"

df_2019=pd.merge(df_deaths_2019, df_happiness_2019, on="Entity")
```

```python
df_2019
```

```python
df_2019 = df_2019.drop(['Code','Cantril ladder score'], axis=1)
```

```python
df_2019['deaths_by_population'] = (df_2019["Total-deaths"]/df_2019['Population (historical estimates)'])*10000
```

```python
df_2019
```

```python
#putting them in descending order according to the total deaths
df_2019_des = df_2019.sort_values(by=['deaths_by_population'], ascending=False)
```

```python
df_2019_des
```

```python
df_2019_very = df_2019_des[0:5]
df_2019_low = df_2019_des[5:len(df_2019_des)]
```

```python
df_2019_very
```

```python
# define the three classes

class_names = ['Low', 'Medium', 'High', 'Very High']
```

```python
# divide the data for deaths by population into 3 equally spaced intervals using 4 numbers starting from the minimum and ending with the maximum

classes = np.linspace(min(df_2019_low["deaths_by_population"]), max(df_2019_low["deaths_by_population"]), 4)
classes
```

```python
classes = np.append(classes, 2.5)
```

```python
classes
```

```python
# add a new column to the dataframe with the class information

df_2019_des['grouped_deaths'] = pd.cut(df_2019_des["deaths_by_population"], classes, labels=class_names, include_lowest=True )
```

```python
# Count the frequency for each class in the newly added column "grouped_deaths"
df_2019_des["grouped_deaths"].value_counts()
```

```python
pyplot.bar(class_names, df_2019_des["grouped_deaths"].value_counts())

plt.xlabel("Deaths")
plt.ylabel("count")
```

```python
sns.boxplot(x="grouped_deaths", y="deaths_by_population", data=df_2019_des)
```

## Mean

```python
# mean for each year
year = 2000
mean_list = []
year_list = []
for i in range(20):
  df_deaths_mean = df_deaths[df_deaths['Year'] == year ]
  mean = df_deaths_mean["Total-deaths"].mean()
  mean_list.append(mean)
  year_list.append(year)
  year = year + 1
plt.plot(year_list, mean_list)
plt.xlabel("Years")
plt.ylabel("Average deaths")
```

## Variance

```python
df_2019
```

```python
# visualize the distribution of the deaths

sns.displot(df_2019['deaths_by_population']);
```

```python
# calculate Skewness of death toll
# if skewness falls between -0.5 and 0.5, the distribution is almost symmetric.
# if skewness falls between -1 and -0.5 or between 0.5 and 1, the distribution is slightly skewed to the left/right.
# if skewness is lower than -1 or bigger than 1 , the distribution is extremely skewed to the left/right.

skewdeath = df_2019['deaths_by_population'].skew(axis = 0, skipna = True)
print('skewness: ', skewdeath)
```

```python
# add a new column called "GDP_binned" to the dataframe.
# "GDP_binned" is created using the five number summary of the initail GDP per capita column
# note that "GDP_binned" contains categorical data

bin_GDP = [0, 5234, 14441, 32868, 120000]
category_GDP = ['<5235', '5235-14442','14442-32869','>32869']
df_2019['GDP_binned'] = pd.cut(df_2019['GDP per capita, PPP (constant 2017 international $)'], bins=bin_GDP, labels=category_GDP)
```

```python
df_2019
```

```python
sns.boxplot(x="GDP_binned", y="deaths_by_population", data=df_2019)
```
